## Regresion lineal

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(X,prepend = True)
```

```python
modelo = sm.OLS(Y,X) ## modelo ordinary list square
modelo = modelo.fit()
```

```python
print(modelo.summary())

## variale dependiente
## R squared 0-1 
##const 
## pt>etc es mejor que este en 0.00 porque asi  no estropea al modelo, en la sig columa todos del mismo sigono
```
