```python
##Analizando limpiando y llenando nuestros daots
#Llenamos los valores nulos con "ND" de los stirngs(en python son objects)
##Llenamos los valores nulos con la media si el campo es de tipo numerico
base["MasVnrType"] = base["MasVnrType"].fillna("ND")
```

```python
base["BsmtQual"] = base["BsmtQual"].fillna("ND")
```

```python
base["BsmtCond"] = base["BsmtCond"].fillna("ND")
```

```python
base["BsmtExposure"] = base["BsmtExposure"].fillna("ND")
```

```python
base["GarageType"] = base["GarageType"].fillna("ND")
```

```python
base["GarageCond"] = base["GarageCond"].fillna("ND")
```

```python
base["GarageYrBlt"] = base["GarageYrBlt"].fillna(base["GarageYrBlt"].mean())
```
