```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
base = pd.read_csv("audi.csv")
```

```python
base.head(10)
```

```python
base.info()
```

```python
base["model"].value_counts()
base["transmission"].value_counts()
```

```python
## hacer un drop en la info que no sirve (fueltype y engine size)
```

```python
base["model"] = pd.factorize(base["model"])[0]
```

```python
base["transmission"] = pd.factorize(base["transmission"])[0]
```

```python
base["fuelType"] = pd.factorize(base["fuelType"])[0]
```

```python
base_num = base.select_dtypes(include=["number"])
```

```python
##Matriz de correlacion
```

```python
corr = base_num.corr()
```

```python
plt.figure(figsize = (32,16))
sn.heatmap(corr,annot=True)
```

```python
corr.price.sort_values(ascending = False)
```

```python
X = base_num[["year","engineSize","model","transmission","tax"]]
```

```python
Y = base_num["price"]
```

```python
base_aux = pd.concat([X,Y],axis=1)
```

```python
plt.figure(figsize=(25,16))
sn.pairplot(base_aux)
```

```python
##Regresion lineal
```

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(X,prepend = True)
```

```python
modelo = sm.OLS(Y,X) ## modelo ordinary list square
modelo = modelo.fit()
```

```python
print(modelo.summary())
```

```python
X.head(1)
```

```python
print(modelo.predict([1.0,2017,1.4,0,0,150]))
```
