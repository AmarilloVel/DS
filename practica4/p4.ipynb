```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import statsmodels.api as sm
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
base = pd.read_csv("Blood Presssure Abnormality.csv")
```

```python
base = base.drop(["Patient_Number"], axis = 1)
```

```python
base.head(50)
```

```python
base.info()
```

```python
base["Pregnancy"] = base["Pregnancy"].fillna(0)
```

```python
base["alcohol_consumption_per_day"] = base["alcohol_consumption_per_day"].fillna((base["alcohol_consumption_per_day"].mean()))
```

```python
base["Genetic_Pedigree_Coefficient"] = base["Genetic_Pedigree_Coefficient"].fillna((base["Genetic_Pedigree_Coefficient"].mean()))
```

```python
base.describe()
```

## Modelo de Regresion Logistica

```python
from sklearn import model_selection
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
y = base["Blood_Pressure_Abnormality"] #Variable dependiente
x = base.drop("Blood_Pressure_Abnormality", axis = 1) # La variables quitando la dependiente
```

```python
y
```

```python
x.head(5)
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y,test_size = 0.2, random_state = 2)
```

```python
x_train.describe()
```

```python
x.describe()
```

```python
y_train.sum()/len(y_train)
```

```python
y_test.sum()/len(y_test)
```

## Cross Validation

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model = LogisticRegression(solver="liblinear")
model.fit(x_train,y_train)
```

```python
predict = model.predict(x_test)
```

```python
# Factor de exito del modelo
score = model.score(x_test,y_test)
print(score)
```

### Evaluacion del modelo

```python
pd.crosstab(y_test,predict)
```

```python
print(classification_report(y_test,predict))
```

```python
probs_df = pd.DataFrame(predict)
```

```python
# Defifinmos un umbral para cambiarlo del 50% a uno mas aproiado al problema de analisis
t= 0.2 # esto significara que todos los pacientes que tengan una probabilidad mayor seran 
# diagnosticados con Anormalidad de la presión arterial
probs_df["Prediccion"] = np.where(probs_df[0] >= t,1,0)
# o sea que si la probabilidad es mayor que 0.2 que le asigne 1 a esta nueva columna
#de lo contrario que le asigne 0
#ahora agregamos una columna mas con el valor real de set de prueba y_test
probs_df["Valor_real"] = list(y_test)
```
