```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
# jalamos los datos 
base = pd.read_csv("./seattle-weather.csv")
```

```python
base.info()
```

```python
base["weather"].value_counts()
```

```python
base.head(50)
```

```python
#Factorizamos los valores de este campo con valores numericos que podamos usar en la ecucaion 
base["weather"] = pd.factorize(base["weather"])[0]
```

```python
##Tempreatura promedio
temp_avg = base[["temp_max","temp_min"]].mean(axis=1)
base["temp_avg"] = temp_avg
```

```python
base.head(50)
```

```python
base["weather"].value_counts()
#0-drizzle
#1-rain
#2-sun
#3-snow
#4-fog
```

```python
base_num = base.select_dtypes(include=["number"])
```

```python
base_num.info()
```

```python
corr = base.corr()
```

```python
plt.figure(figsize = (32,16))
sn.heatmap(corr,annot=True)
```
